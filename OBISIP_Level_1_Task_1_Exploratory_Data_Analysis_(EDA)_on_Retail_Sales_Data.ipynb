```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
%matplotlib inline
```

```python
from google.colab import files
uploaded = files.upload()
df = pd.read_csv(list(uploaded.keys())[0])
print("✅ Data loaded:", df.shape)
```

```python
display(df.head(), df.info(), df.describe(include='all'), df.isnull().sum())
```

```python
dups = df.duplicated().sum()
print("Duplicates:", dups)
df.drop_duplicates(inplace=True)
```

```python
missing = df.isnull().sum()
print("Missing values:\n", missing)
# Example: If 'Sales' has nulls, drop them or fill with mean:
# df['Sales'].fillna(df['Sales'].mean(), inplace=True)
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    vals = df[col].dropna()
    print(f"\n-- {col} stats --")
    print("Mean:", np.mean(vals))
    print("Median:", np.median(vals))
    print("Mode:", stats.mode(vals, keepdims=True).mode[0])
    print("Std Dev:", np.std(vals, ddof=1))
```

```python
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
    monthly = df.resample('M')['Sales'].sum()
    plt.figure(figsize=(12,5))
    monthly.plot(title='Monthly Sales Trend', marker='o')
    plt.ylabel('Sales')
    plt.show()
```

```python
if 'CustomerID' in df.columns:
    top_cust = df.groupby('CustomerID')['Sales'].sum().sort_values(ascending=False).head(10)
    plt.figure(figsize=(8,4))
    top_cust.plot(kind='bar', title='Top 10 Customers by Sales')
    plt.ylabel('Total Sales')
    plt.show()
```

```python
if 'Product' in df.columns:
    top_prod = df.groupby('Product')['Quantity'].sum().sort_values(ascending=False).head(10)
    plt.figure(figsize=(8,4))
    top_prod.plot(kind='bar', title='Top 10 Products by Quantity Sold')
    plt.ylabel('Units Sold')
    plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation')
plt.show()
```

```python
if 'Region' in df.columns:
    df_reset = df.reset_index()
    pivot = df_reset.pivot_table(values='Sales', index=df_reset['Date'].dt.to_period('M'), columns='Region', aggfunc='sum')
    pivot.plot(figsize=(12,6), title='Monthly Sales by Region')
    plt.xlabel('Month'); plt.ylabel('Sales')
    plt.show()
```
